```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('/content/Weather Dataset - Trial Activity DataSet.csv.csv')
df.head()
```

```python
sns.barplot(x='weather_description',y='humidity',data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.swarmplot(x="weather_type",y="wind_direction",data=df,palette="winter")
```

```python
sns.countplot(x="weather_type",data=df,palette='viridis')
```

```python
sns.distplot(df['humidity'])
```

```python

sns.pointplot(x="air_pollution_index",y="wind_speed",hue="weather_type", data=df,palette="autumn")
```

```python
sns.pairplot(df)
```

```python
sns.jointplot(x='dew_point',y="humidity",hue="dew_point",data=df,palette="summer")
```

```python
sns.lmplot(x="dew_point",y="humidity",hue="dew_point",data=df)
```

```python
sns.stripplot(x="dew_point",y="visibility_in_miles",data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.boxplot(x="dew_point",data=df)
```
